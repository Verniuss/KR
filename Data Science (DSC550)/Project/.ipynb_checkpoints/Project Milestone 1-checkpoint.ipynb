```python
## Looking at fnlwgt of Male & Female. How it distributes separately

g = sns.FacetGrid(df, col = 'Sex')

g.map_dataframe(sns.histplot, 'BMI');
```

```python
# What I was trying to do here is separate everything into desireable categories 
# and look for any trends or commonalities. However The Scatter plot wasn't very useful here to visualize the data

g = sns.FacetGrid(df, col = 'Sex', row = 'AgeCategory', hue = 'HeartDisease', sharey = False)

g.map_dataframe(sns.scatterplot, x = 'Diabetic', y = 'BMI')
g.set_axis_labels('Diabetic', 'BMI')
g.set_titles(col_template = '{col_name}', row_template = '{row_name}' )
g.add_legend();
```

```python
# I used a different X (General Health) because I thought that this was more useful 
# than just checking for diabetes. I could do a diabetes comparison in a later plot set. 
# But here we can get a general idea of how well the health is of these individuals with the given conditions.
g = sns.FacetGrid(df, col = 'Sex', row = 'AgeCategory', hue = 'HeartDisease', sharey = False)

g.map_dataframe(sns.histplot, x = 'GenHealth', y = 'BMI')
g.set_axis_labels('GenHealth', 'BMI')
g.set_titles(col_template = '{col_name}', row_template = '{row_name}' )
g.add_legend();
```
