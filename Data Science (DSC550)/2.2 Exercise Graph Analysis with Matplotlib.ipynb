```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('C:/Users/Kyle/Documents/Data Science (DSC550)/2.2 Exercise/new_cars_prices.csv')
```

```python
# Filter the data for Hyundai car models
hyundai_df = df[df['Car Model'].str.contains('Hyundai')]

# Sort the data by new price in descending order and select top 10 rows
top_10 = hyundai_df.sort_values(by='New Price', ascending=False).head(10)

# Histogram for the distribution of new prices for top 10 Hyundai car models
sns.histplot(x='New Price', data=top_10, bins=20, kde=False)
sns.set_style('whitegrid')
plt.title('Distribution of New Prices for Top 10 Hyundai Car Models')
plt.xlabel('New Price')
plt.ylabel('Frequency')
plt.show()
```

```python
#The histogram shows the distribution of new prices for Hyundai car models. 
#The x-axis represents the new prices of the cars, and the y-axis shows the frequency 
#of occurrence for each price range. The bins are set to 20, which means the range of prices 
#is divided into 20 equal intervals. The height of each bar represents the number of cars that 
#fall within that price range. From the graph, it can be seen that the majority of the Hyundai 
#car models have new prices between approximately 500,000 and 1,000,000 EGP. The graph is skewed 
#to the right, indicating that there are fewer car models with higher prices. The absence of a kde 
#line suggests that the distribution of new prices for Hyundai car models is not continuous, but rather discrete.
```

```python
# Sort the data by new price in descending order and select top 10 rows
top_10 = df.sort_values(by='New Price', ascending=False).head(10)

# Horizontal bar chart
plt.barh(y=top_10['Car Model'], width=top_10['New Price'])
plt.title('Top 10 Cars with the Highest New Price')
plt.xlabel('New Price')
plt.ylabel('Car Model')
plt.show()
```

```python
#The horizontal bar chart is showing the top 10 cars with the highest new price in the dataset. 
#The cars are listed on the y-axis and their corresponding new prices are shown on the x-axis. 
#The length of each bar represents the price of the car, with longer bars indicating higher prices.

#From the chart, we can see that the Bugatti Chiron Sport 110 Ans and the Rolls Royce 
#Phantom are the two most expensive cars in the dataset, with new prices exceeding 20 million EGP. 
#The remaining top 10 cars are also luxury vehicles from brands like Lamborghini, Bentley, and Ferrari, 
#with new prices ranging from 6 million EGP to 14 million EGP.
```

```python
# Filter for Toyota Corolla Automatic / SMART / Sun Roof 2022 and Lada Granta manual / salon 2022
toyota = df[df["Car Model"] == "Toyota Corolla Automatic / SMART / Sun Roof 2022"]
lada = df[df["Car Model"] == "Lada Granta manual / salon 2022"]

# Limit data to top 100 results
toyota = toyota.head(100)
lada = lada.head(100)

# Create scatter plot
sns.scatterplot(x="Old Price", y="New Price", data=toyota, label="Toyota Corolla Automatic / SMART / Sun Roof 2022")
sns.scatterplot(x="Old Price", y="New Price", data=lada, label="Lada Granta manual / salon 2022")

# Add title and axis labels
plt.title("Comparison of Toyota Corolla Automatic / SMART / Sun Roof 2022 and Lada Granta manual / salon 2022")
plt.xlabel("Old Price")
plt.ylabel("New Price")

# Show the plot
plt.show()
```

```python
#The scatter plot compares the cost of two car models: Toyota Corolla Automatic / SMART / Sun Roof 2022 and Lada Granta manual 
#/ salon 2022. Each point on the plot represents a specific dealership and its respective price for the car model. 
#The x-axis represents the price of the Toyota Corolla while the y-axis represents the price of the Lada Granta.

#We can see that the majority of the data points cluster towards the bottom left corner, 
#indicating that the Lada Granta tends to be less expensive than the Toyota Corolla. However, 
#there are some data points that are outliers and fall outside of the main cluster, indicating 
#that some dealerships may be pricing the cars differently. Additionally, we can see that there are 
#fewer data points for the Lada Granta than for the Toyota Corolla, which may suggest that the 
#Toyota Corolla is a more popular or widely available car model.
```
