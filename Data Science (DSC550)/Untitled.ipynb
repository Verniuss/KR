```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV file into Pandas dataframe
df = pd.read_csv('https://drive.google.com/uc?id=11HfxsAtHNi_JVlB7SO1aGVc240C4mOiq')
```

```python
#This dataset contains official price information for major vegetables 
#and fruits in Nepal from 2013 to 2021. The dataset includes daily price data 
#for each vegetable and fruit, as well as the maximum, minimum, and average 
#prices over the period. The prices are based on official figures and provide a 
#valuable resource for anyone interested in analyzing the prices of agricultural commodities in Nepal.
```

```python
# Filter data for Tomato Big (Nepali)
tomato_df = df[df['Commodity'] == 'Tomato Big(Nepali)']
```

```python
df.head()
```

```python
# convert date column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
```

```python
# create a new column with the month and year information
df['MonthYear'] = df['Date'].dt.strftime('%Y-%m')
```

```python
# group the data by month and year and calculate the average price for each commodity
grouped = df.groupby(['MonthYear', 'Commodity']).mean().reset_index()

# create a pivot table to plot the data
pivot = grouped.pivot(index='MonthYear', columns='Commodity', values='Average')

# plot the pivot table
fig, ax = plt.subplots(figsize=(15, 10))
pivot.plot(ax=ax)

# set the title and labels
plt.title('Average Vegetable Prices in Nepal from June 2013 to July 2015')
plt.xlabel('Month/Year')
plt.ylabel('Price (NPR)')

# adjust the x-axis labels
plt.xticks(rotation=45, ha='right', fontsize=8)

# show the plot
plt.show()
```
